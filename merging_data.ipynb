```python
datamr['skills'] = datamr['skills'].str.strip("[]")  # Remove square brackets
datamr['skills'] = datamr['skills'].str.replace("\\n\\nLet me know if you need anything else!", "", regex=False)  # Remove the unwanted text
datamr['skills'] = datamr['skills'].str.replace("'", "")  # Remove single quotes
datamr['skills'] = datamr['skills'].str.replace("!", "")  # Remove single quotes
datamr['skills'] = datamr['skills'].str.replace("(", "") # Remove single quotes
datamr['skills'] = datamr['skills'].str.replace(")", "") # Remove single quotes
datamr['skills'] = datamr['skills'].str.replace("\\n\\nLet me know if you need any further assistance", "", regex=False)  # Remove the second unwanted text
datamr['skills'] = datamr['skills'].str.split(", ")  # Split by comma and space

# Flatten the lists and extract unique skills
all_skills = [skill for sublist in datamr['skills'] if isinstance(sublist, list) for skill in sublist if skill]  # Check if it's a valid list
unique_skills = list(set(all_skills))

# Display the unique skills
print(unique_skills)
```

```python

# 2. Renommer la colonne 'title' en 'job_title'
datamr.rename(columns={'title': 'job_title'}, inplace=True)

# 1. Ajouter une colonne 'salary' avec des valeurs 0 entre 'location' et 'sector'
datamr['salary'] = 0
# Réorganiser les colonnes pour placer 'salary' entre 'location' et 'sector'
cols = list(datamr.columns)
location_index = cols.index('location')
# Insérer 'salary' à l'index approprié
cols.insert(location_index + 1, cols.pop(cols.index('salary')))
datamr = datamr[cols]
```

```python
# 3. Placer la colonne 'job_title' avant 'skills'
job_title_index = cols.index('job_title')
skills_index = cols.index('skills')
# Réorganiser les colonnes pour placer 'job_title' avant 'skills'
if job_title_index > skills_index:
    cols.insert(skills_index, cols.pop(job_title_index))
datamr = datamr[cols]

# 4. Ajouter trois colonnes à valeurs 0: 'day', 'month', 'year'
datamr['day'] = 0
datamr['month'] = 0
datamr['year'] = 0
```

```python
# Créer des tables de correspondance (sous forme de dictionnaires)
companies_dict = {name: idx + 1 for idx, name in enumerate(data['companyName'].unique())}
contract_types_dict = {name: idx + 1 for idx, name in enumerate(data['contractType'].unique())}
experience_levels_dict = {name: idx + 1 for idx, name in enumerate(data['experienceLevel'].unique())}
locations_dict = {name: idx + 1 for idx, name in enumerate(data['location'].unique())}
sectors_dict = {name: idx + 1 for idx, name in enumerate(data['sector'].unique())}
skills_dict = {name: idx + 1 for idx, name in enumerate(data['skills'].unique())}
titles_dict = {name: idx + 1 for idx, name in enumerate(data['job_title'].unique())}
salary_dict = {name: idx + 1 for idx, name in enumerate(data['salary'].unique())}
country_dict = {name: idx + 1 for idx, name in enumerate(data['country'].unique())}

# Remplacer les valeurs dans le DataFrame par les IDs
data['companyName'] = data['companyName'].map(companies_dict)
data['contractType'] = data['contractType'].map(contract_types_dict)
data['experienceLevel'] = data['experienceLevel'].map(experience_levels_dict)
data['location'] = data['location'].map(locations_dict)
data['sector'] = data['sector'].map(sectors_dict)
data['skills'] = data['skills'].map(skills_dict)
data['job_title'] = data['job_title'].map(titles_dict)
data['salary'] = data['salary'].map(salary_dict)
data['country'] = data['country'].map(country_dict)
data
```

```python
for company_id, contract_id, experience_id , location_id , salary_id, sector_id, title_id, skills_id,country_id, date_id in zip(data['companyName'],data['contractType'],data['experienceLevel'],data['location'],data['salary'],data['sector'],data['job_title'],data['skills'],data['country'],data['date']):
    cursor.execute("INSERT INTO jobs (company_id,contract_type_id,experience_level_id,location_id,salary_id,sector_id,title_id,skill_id,country_id,date_id)  VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)",(company_id, contract_id, experience_id , location_id ,salary_id, sector_id,title_id, skills_id,country_id,date_id,))
connection.commit()
```
